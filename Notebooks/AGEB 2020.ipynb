```python
DEUNE_20 = pd.read_csv('Datos/denue_21_0420_csv/conjunto_de_datos/denue_inegi_21_.csv', delimiter=',',encoding='latin1',low_memory=False)
```

```python
DEUNE_20
```

```python
DEUNE_20=DEUNE_20[['ageb','codigo_act','per_ocu']]
DEUNE_20.groupby('ageb')
```

```python
agebs_deune = set(DEUNE_20['ageb'].unique())
faltantes = agebs_cpv - agebs_deune
print(faltantes)
```

```python
DEUNE_20 = DEUNE_20[DEUNE_20['ageb'].isin(agebs_cpv)]
DEUNE_20 = DEUNE_20.sort_values('ageb')
DEUNE_20
```

```python
DEUNE_20['per_ocu'].unique()
```

```python
reemplazos = {
    '0 a 5 personas': 2,
    '6 a 10 personas': 8,
    '11 a 30 personas': 20,
    '31 a 50 personas': 40,
    '51 a 100 personas': 75,
    '101 a 250 personas': 175,
    '251 y más personas': 300
}

DEUNE_20['per_ocu'] = DEUNE_20['per_ocu'].replace(reemplazos)
DEUNE_20
```

```python
# Convertir los valores de la columna 'per_ocu' a cadenas y extraer los primeros dos dígitos
DEUNE_20['primeros_dos_digitos'] = DEUNE_20['codigo_act'].astype(str).str[:2]

# Crear una columna para cada posible valor de los primeros dos dígitos
# Esto va a crear columnas para cada valor único de los primeros dos dígitos
DEUNE_20_wide_1 = DEUNE_20.pivot_table(index='ageb', columns='primeros_dos_digitos', values='codigo_act', aggfunc='count')
DEUNE_20_wide_2 = DEUNE_20.pivot_table(index='ageb', columns='primeros_dos_digitos', values='per_ocu', aggfunc='mean')

DEUNE_20_wide_1['31-33'] = DEUNE_20_wide_1[['31', '32', '33']].sum(axis=1)
DEUNE_20_wide_1 = DEUNE_20_wide_1.drop(columns=['31', '32', '33'])

DEUNE_20_wide_2['31-33'] = DEUNE_20_wide_2[['31', '32', '33']].mean(axis=1)
DEUNE_20_wide_2 = DEUNE_20_wide_2.drop(columns=['31', '32', '33'])

DEUNE_20_wide_1['48-49'] = DEUNE_20_wide_1[['48', '49']].sum(axis=1)
DEUNE_20_wide_1 = DEUNE_20_wide_1.drop(columns=['48', '49'])

DEUNE_20_wide_2['48-49'] = DEUNE_20_wide_2[['48', '49',]].mean(axis=1)
DEUNE_20_wide_2 = DEUNE_20_wide_2.drop(columns=['48', '49'])

# Renombrar las columnas agregando un prefijo "grupo_"
DEUNE_20_wide_1.columns = [f"grupo_{col}" for col in DEUNE_20_wide_1.columns]
DEUNE_20_wide_2.columns = [f"per_ocu_{col}" for col in DEUNE_20_wide_2.columns]


# Unir las columnas de vuelta al DataFrame original si lo necesitas
DEUNE_20_final = pd.concat([DEUNE_20_wide_1,DEUNE_20_wide_2], axis=1)

# Ver el resultado
DEUNE_20_final=DEUNE_20_final.fillna(0)
DEUNE_20_final
```

```python
AGEB_2020 = AGEB_2020.join(DEUNE_20_final, how='left')
AGEB_2020
```
